## Notion de signal

Un *signal* est un vecteur $x \in \mathbb{R}^N$ représentant une séquence de coefficients appelés échantillons. $x[n]$ est l'amplitude du $n$-ième échantillon. Exemple, un son est un signal.

La $fréquence$ d'échantillonage (sampling frequency) est le nombre d'échantillons par seconde.

Exemple : $f_1 = 44 100 Hz$ pour un CD
